```python
import scanpy as sc
import pandas as pd
import os
from starcat import starCAT
```

```python
custom_tcat = starCAT(
    reference="/sibcb1/bioinformatics/yangyue/project/immunotherapy/7.2.cNMF_CD8T/Example_refbuilder_95_v2/starcat_refstarcat_consensus_spectra_normalized.filtered.txt"
)

input_dir = "/sibcb1/bioinformatics/yangyue/project/immunotherapy/1.new_data_h5ad_v2/"
output_dir = "/sibcb1/bioinformatics/yangyue/project/immunotherapy/7.2.cNMF_CD8T/4.3.CD8T_GEP_usage"
os.makedirs(output_dir, exist_ok=True)

h5ad_files = sorted([
    f for f in os.listdir(input_dir)
    if f.endswith("_CD8T.h5ad")
])

print(f"Found {len(h5ad_files)} CD8T datasets")
```

```python
for fname in h5ad_files:
  dataset_name = fname.replace("_CD8T.h5ad", "")
  print(f"\n=== Processing {dataset_name} ===")
  
  # 1. 读取数据
  adata = sc.read_h5ad(os.path.join(input_dir, fname))
  
  # 2. 检查 celltype
  if 'celltype' not in adata.obs.columns:
    raise ValueError(f"{dataset_name}: obs 中不存在 celltype 列")
  
  query_adata = adata[adata.obs['celltype'] == 'CD8T'].copy()
  print(f"_CD8T cells: {query_adata.n_obs}")
  
  # 修正位置：continue 必须缩进！
  if query_adata.n_obs == 0:
    print(f"⚠️ No CD8T cells in {dataset_name}, skip.")
    continue 
  
  # 3. starCAT 投影 (现在可以执行到了)
  usage, _ = custom_tcat.fit_transform(query_adata)
  
  # 4. 保存 cell-level usage
  usage_path = os.path.join(
    output_dir,
    f"{dataset_name}_cGEP_usage_cell_level.csv"
  )
  usage.to_csv(usage_path)
  
  print(f"Saved → {usage_path}")
```
